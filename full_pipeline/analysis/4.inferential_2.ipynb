```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

def final_inferential_analysis():
    """
    Melakukan analisis inferensial gabungan:
    1. Dwell time dalam detik.
    2. Analisis alur perilaku pelanggan yang komprehensif.
    """
    input_path = os.path.join('preprocessed_data', 'cleaned_action_shelf_log.csv')
    # Buat direktori output baru untuk analisis final yang paling komprehensif
    output_dir = 'inferential_outputs_final'

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    print(f"Membaca data bersih dari: {input_path}")
    try:
        df = pd.read_csv(input_path)
    except FileNotFoundError:
        print(f"Error: File tidak ditemukan di {input_path}")
        print("Jalankan script '01_preprocessing.py' terlebih dahulu.")
        return

    # --- Analisis 1: Dwell Time (dalam Detik) ---
    print("\n--- Memulai Analisis Dwell Time ---")
    fps = 13.0
    df['next_shelf'] = df.groupby('pid')['shelf_id'].shift(-1)
    dwell_time_data = []
    for pid, group in df.groupby('pid'):
        start_frame, current_shelf = 0, -1
        for i, row in group.iterrows():
            if row['shelf_id'] != current_shelf:
                if current_shelf != -1:
                    dwell_seconds = (end_frame - start_frame) / fps
                    dwell_time_data.append({'pid': pid, 'shelf_id': current_shelf, 'dwell_time': dwell_seconds})
                current_shelf, start_frame = row['shelf_id'], row['frame']
            end_frame = row['frame']
        if current_shelf != -1:
            dwell_seconds = (end_frame - start_frame) / fps
            dwell_time_data.append({'pid': pid, 'shelf_id': current_shelf, 'dwell_time': dwell_seconds})
    
    dwell_df = pd.DataFrame(dwell_time_data)
    avg_dwell_time = dwell_df.groupby('shelf_id')['dwell_time'].mean().sort_values(ascending=False)
    avg_dwell_time.to_csv(os.path.join(output_dir, '1_average_dwell_time_per_shelf.csv'))
    print("Analisis Dwell Time selesai. Hasil tabel disimpan.")
    
    plt.figure(figsize=(12, 7))
    sns.barplot(x=avg_dwell_time.index, y=avg_dwell_time.values, palette='plasma', order=avg_dwell_time.index)
    plt.title('Rata-rata Dwell Time (dalam Detik) per Rak', fontsize=16)
    plt.xlabel('ID Rak (shelf_id)', fontsize=12)
    plt.ylabel('Rata-rata Dwell Time (Detik)', fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, '1_average_dwell_time_chart.png'))
    print("Grafik Dwell Time disimpan.")
    plt.show()
    plt.close()

    # --- Analisis 2: Alur Perilaku Pelanggan Komprehensif ---
    print("\n--- Memulai Analisis Alur Perilaku Komprehensif ---")

    # 1. Identifikasi semua interaksi unik (pelanggan, rak)
    interactions = df[['pid', 'shelf_id']].drop_duplicates()

    # 2. Tandai setiap interaksi dengan aksi kunci yang terjadi
    reach_events = df[df['action'] == 'Reach To Shelf'][['pid', 'shelf_id']].drop_duplicates().assign(did_reach=True)
    inspect_events = df[df['action'] == 'Inspect Product'][['pid', 'shelf_id']].drop_duplicates().assign(did_inspect=True)
    return_events = df[df['action'] == 'Hand In Shelf'][['pid', 'shelf_id']].drop_duplicates().assign(did_return=True)

    # 3. Gabungkan semua informasi ke dalam satu tabel analisis
    analysis_df = pd.merge(interactions, reach_events, on=['pid', 'shelf_id'], how='left')
    analysis_df = pd.merge(analysis_df, inspect_events, on=['pid', 'shelf_id'], how='left')
    analysis_df = pd.merge(analysis_df, return_events, on=['pid', 'shelf_id'], how='left')
    analysis_df = analysis_df.fillna(False)

    # 4. Kategorikan hasil akhir dari setiap interaksi
    def categorize_outcome(row):
        if not row['did_reach']:
            return 'No Reach' # Abaikan interaksi yang tidak dimulai dengan jangkauan
        
        if row['did_inspect'] and row['did_return']:
            return 'Keraguan & Pembatalan'
        elif row['did_inspect'] and not row['did_return']:
            return 'Konversi Sukses'
        else: # did_reach is True, but did_inspect is False
            return 'Kegagalan Menarik Minat'

    analysis_df['outcome'] = analysis_df.apply(categorize_outcome, axis=1)
    
    relevant_outcomes = analysis_df[analysis_df['outcome'] != 'No Reach']

    # 5. Agregasi hasil per rak dan hitung persentasenya
    outcome_summary = relevant_outcomes.groupby(['shelf_id', 'outcome']).size().unstack(fill_value=0)
    outcome_percentage = outcome_summary.div(outcome_summary.sum(axis=1), axis=0) * 100
    
    desired_order = ['Konversi Sukses', 'Keraguan & Pembatalan', 'Kegagalan Menarik Minat']
    for col in desired_order:
        if col not in outcome_percentage.columns:
            outcome_percentage[col] = 0
    outcome_percentage = outcome_percentage[desired_order]
    
    outcome_percentage = outcome_percentage.sort_values(by='Konversi Sukses', ascending=False)

    outcome_percentage.to_csv(os.path.join(output_dir, '2_comprehensive_journey_summary.csv'))
    print("Analisis Alur Komprehensif selesai. Hasil tabel disimpan.")

    # 6. Visualisasi sebagai 100% Stacked Bar Chart
    plt.style.use('seaborn-v0_8-whitegrid')
    outcome_percentage.plot(
        kind='bar', 
        stacked=True, 
        figsize=(14, 8),
        color=['#2ca02c', '#ff7f0e', '#d62728'] # Hijau (Sukses), Merah (Batal), Oranye (Gagal)
    )

    plt.title('Analisis Alur Perilaku Pelanggan per Rak', fontsize=18, pad=20)
    plt.xlabel('ID Rak (shelf_id)', fontsize=12)
    plt.ylabel('Persentase Hasil Interaksi (%)', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(range(0, 101, 10))
    plt.legend(title='Hasil Akhir Interaksi', bbox_to_anchor=(1.02, 1), loc='upper left')
    plt.tight_layout(rect=[0, 0, 0.85, 1])
    
    plt.savefig(os.path.join(output_dir, '2_comprehensive_journey_chart.png'))
    print("Grafik Alur Komprehensif disimpan.")
    plt.show()
    plt.close()

    print(f"\nAnalisis Inferensial FINAL Selesai. Hasil disimpan di folder '{output_dir}'.")

if __name__ == '__main__':
    final_inferential_analysis()
```

```python
print("\n--- Membuat Tabel Ringkasan dengan Arketipe Perilaku ---")

output_dir = 'inferential_outputs_final'

# 1. Muat data yang sudah dihitung sebelumnya
df_raw = pd.read_csv(os.path.join('preprocessed_data', 'cleaned_action_shelf_log.csv'))
df_dwell = pd.read_csv(os.path.join(output_dir, '1_average_dwell_time_per_shelf.csv'))
df_outcomes = pd.read_csv(os.path.join(output_dir, '2_comprehensive_journey_summary.csv')).set_index('shelf_id')

# 2. Hitung jumlah interaksi unik
unique_interactions = df_raw.groupby('shelf_id')['pid'].nunique().reset_index()
unique_interactions.rename(columns={'pid': 'Interaksi Unik'}, inplace=True)

# 3. Gabungkan data dasar
summary_table = pd.merge(unique_interactions, df_dwell, on='shelf_id')
summary_table = summary_table.set_index('shelf_id')

# 4. Fungsi untuk menentukan Arketipe Perilaku (Versi Final)
def get_behavioral_archetype(row):
    shelf_id = row.name
    dwell_time = row['dwell_time']
    unique_visits = row['Interaksi Unik']
    
    # Cek apakah rak ada di data outcomes (artinya ada interaksi fisik 'Reach')
    if shelf_id not in df_outcomes.index:
        if dwell_time > 3.0:
            return 'Passive Attention (No Physical Engagement)'
        else:
            return 'Low Engagement Zone'
            
    outcomes = df_outcomes.loc[shelf_id]
    dominant_outcome = outcomes.idxmax()
    
    # Arketipe paling berharga: ada sinyal konversi
    if outcomes.get('Konversi Sukses', 0) > 10:
        return 'High Attention, Low Conversion'
        
    # Arketipe kedua: ada interaksi mendalam tapi batal
    if dominant_outcome == 'Keraguan & Pembatalan':
        return 'Interaksi Positif Namun Ragu'
        
    # Arketipe ketiga: banyak dilewati tapi diabaikan
    if dominant_outcome == 'Kegagalan Menarik Minat':
        if unique_visits > 8:
            return 'Traffic Tinggi, Engagement Rendah'
        else:
            return 'Low Engagement Zone'
            
    return 'Lainnya' # Fallback

# 5. Terapkan fungsi untuk membuat kolom Arketipe
summary_table['Arketipe Perilaku'] = summary_table.apply(get_behavioral_archetype, axis=1)

# 6. Format tabel akhir
summary_table.reset_index(inplace=True)
summary_table.rename(columns={'shelf_id': 'Rak', 'dwell_time': 'Rata-rata Dwell (s)'}, inplace=True)
summary_table['Rata-rata Dwell (s)'] = summary_table['Rata-rata Dwell (s)'].round(2)
summary_table = summary_table.sort_values(by='Interaksi Unik', ascending=False).set_index('Rak')

# Simpan ke CSV
summary_table.to_csv(os.path.join(output_dir, '3_behavioral_archetype_summary.csv'))

print("Tabel ringkasan arketipe perilaku berhasil dibuat dan disimpan.")
print(summary_table[['Interaksi Unik', 'Rata-rata Dwell (s)', 'Arketipe Perilaku']])
```
